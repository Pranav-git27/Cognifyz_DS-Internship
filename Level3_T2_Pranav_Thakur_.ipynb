Task 2 Task: Customer Preference Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load data
df = pd.read_csv('/content/Dataset .csv')

# basic info
df.info()
```

```python
# drop missing values
df_cleaned = df.dropna(subset=['Cuisines']).copy()

# split and explode the Cuisines column
df_exploded = df_cleaned.assign(Cuisines=df_cleaned['Cuisines'].str.split(', ')).explode('Cuisines')

print(f"Original rows: {len(df)}")
print(f"Rows after explosion: {len(df_exploded)}")

# display first few rows of exploded data
df_exploded[['Restaurant Name','Cuisines','Aggregate rating','Votes']].head()
```

1.Analyze the relationship between the type of cuisine and the restaurant's rating.

```python

# calculate statistics for each Cuisine
cuisine_stats = df_exploded.groupby('Cuisines').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum',
    'Restaurant ID': 'count'
}).rename(columns={
    'Aggregate rating': 'Average Rating',
    'Votes': 'Total Votes',
    'Restaurant ID': 'Restaurant Count'
})

print("Cuisine Statistics calculated successfully.")
```

2.Identify the most popular cuisines among customers based on the number of votes.

```python
# sort by total votes
most_popular = cuisine_stats.sort_values(by='Total Votes', ascending=False).head(10)
print("Top 10 Most Popular Cuisines by Total Votes:")
print(most_popular[['Total Votes', 'Average Rating']])
```

3.Determine if there are any specific cuisines that tend to receive higher ratings.

```python
# filter for cuisines with 10 restaurants and sort by rating
high_rated=cuisine_stats[cuisine_stats['Restaurant Count'] >= 10].sort_values(by='Average Rating', ascending=False).head(10)

print("Top 10 Cuisines with the Highest Average Ratings (Min 10 Restaurants):")
print(high_rated[['Average Rating', 'Restaurant Count']])
```
